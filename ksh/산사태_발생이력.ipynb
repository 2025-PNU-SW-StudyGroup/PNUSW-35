# 산사태 발생 이력

```python
import pandas as pd
```

```python
df = pd.read_csv('./Data/산림청_최근 5년간 전국 산사태 발생 이력_20231114.csv', encoding='cp949')
```

```python
df
```

```python
df['주소'] = df[['상세주소', '상세주소.1', '상세주소.2', '상세주소.3']].astype(str).agg(' '.join, axis=1)
```

```python
df['상세주소'].value_counts()
```

```python
df['주소']
```

```python
df
```

```python
df = df[df['상세주소']=='경상북도']
```

```python
ri_counts = df['주소'].value_counts().reset_index()
ri_counts.columns = ['리', '산사태_발생건수']
```

```python
ri_counts
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Malgun Gothic'   # 윈도우
plt.rcParams['axes.unicode_minus'] = False      # 마이너스 깨짐 방지
```

```python
import geopandas as gpd
```

```python
gdf = gpd.read_file('./Data/LX법정구역경계_리_전국/RI.shp')
```

```python
gdf.head(-10)
```

```python
gdf_gyeongbuk = gdf[
    gdf['adzn_nm'].str.contains('경상북도') & 
    ~gdf['adzn_nm'].str.contains('울릉군')
]
```

```python
merged = gdf_gyeongbuk.merge(ri_counts, left_on='adzn_nm', right_on='리', how='left')
```

```python
merged['산사태_발생건수'] = merged['산사태_발생건수'].fillna(0)
```

```python
merged
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
merged.plot(column='산사태_발생건수',
             cmap='OrRd',             # 붉은색 계열
             linewidth=0.2,
             edgecolor='black',
             legend=True,
             ax=ax)
plt.title('경상북도 리별 산사태 발생 건수', fontsize=15)
plt.axis('off')
plt.show()
```
